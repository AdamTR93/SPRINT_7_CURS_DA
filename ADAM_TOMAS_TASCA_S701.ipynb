Calculadora de l'índex de massa corporal

Escriu una funció que calculi l'IMC ingressat per l'usuari/ària, és a dir, qui ho executi haurà d'ingressar aquestes dades.


```python
#Creamos las medidas que introducirá el usuario y la operación del IMC

peso = float(input('Pon tu peso en Kilogramos (Kg): '))
print('Tu peso es de', + peso ,'Kg')
altura = float(input('Pon tu altura en metros (m): '))
print('Tu atura es de', + altura, 'm')

def calculate_imc(peso,altura):
    IMC = peso/(altura**2)

    print(f'Tu IMC es: {IMC:.2f}')

    #Verifica el IMC y muestra el rango correspondiente

    if IMC < 18.5:
        print('Estas bajo de peso')
    elif 18.5<=IMC<=24.9:
        print('Tu peso es normal')
    elif 25<=IMC<=29.9: 
        print('Tienes sobrepeso')
    else:
        print('Tienes obesidad')
        
calculate_imc(peso,altura)
```

**EXERCICI 3**

Comptador de paraules d'un text.

Escriu una funció que donat un text, mostri les vegades que apareix cada paraula.

```python
from collections import Counter

text = input("escribe tu texto: \n")
signs = ".,?!¿¡()"
translator = str.maketrans(signs, " " * len(signs))
text = text.translate(translator).lower()
words = Counter(text.split())
for palabra,numero in words.items():
    
    print(f'{palabra} : {numero}')
    
```

*Manera de hacer el ejercicio sin bibliotecas*


```python
# Creamos una función para borrar todos los signos de puntuación del texto introducido
def remover_puntuacion(texto):
    signos_puntuacion = [".", ",", "?", "¿", "!", "¡", "(", ")", ";", ":", "-", "'", '"']
    for signo in signos_puntuacion:
        texto = texto.replace(signo, "") # Reemplazamos el signo por 'nada' para que lo borre
    return texto

# Definición de la función para contar la frecuencia de las palabras en el texto
def contador_veces(texto):
    texto = remover_puntuacion(texto)
    lista_palabras = texto.lower().split()
    frecuencia = {}
    for palabra in lista_palabras:
        frecuencia[palabra] = lista_palabras.count(palabra)
    return frecuencia

texto = input('Escribe un texto: ')
resultado = contador_veces(texto)

# Iterar sobre el diccionario de resultados e imprimir cada palabra con su frecuencia por separado
for palabra,numero in resultado.items():
    print(f'{palabra}:{numero}')
```

**EXERCICI 4**

Diccionari invers.

Resulta que el client té una enquesta molt antiga que s'emmagatzema en un diccionari i els resultats els necessita al revés, és a dir, intercanviats les claus i els valors. Els valors i claus en el diccionari original són únics; si aquest no és el cas, la funció hauria d'imprimir un missatge d'advertiment.

```python
diccionario = {1: 'a',
               2: 'b', 
               3: 'c', 
               4: 'b'}  

# Primero hacemos una lista con los valores del diccionario.
valores = list(diccionario.values())

# Luego hacemos un set de valores porque al hacer set, si hay valores duplicados automáticamente los borra.
valores_unicos = set(valores)

# Comparamos cuántos valores tiene uno y otro, si son diferentes quiere decir que el set ha borrado valores duplicados.
if len(valores) != len(valores_unicos):
    print('¡Error! En el diccionario tienes valores duplicados')

# Diccionario invertido sin incluir valores duplicados
diccionario_invertido = {}
valores_vistos = set()

for clave, valor in diccionario.items():
    if valor in valores_vistos:
        print(f'El valor "{valor}" de la clave "{clave}" es duplicado, no se incluirá en el diccionario invertido.')
    else:
        diccionario_invertido[valor] = clave
        valores_vistos.add(valor)

print(diccionario_invertido)
```
