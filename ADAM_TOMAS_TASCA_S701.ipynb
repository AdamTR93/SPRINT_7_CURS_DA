Calculadora de l'índex de massa corporal

Escriu una funció que calculi l'IMC ingressat per l'usuari/ària, és a dir, qui ho executi haurà d'ingressar aquestes dades.


```python

# Definimos una funcion en la cual el usuario añade su peso y su altura. Reemplaza las comas por puntos por si el usuario lo escribe mal.
# Seguidamente, realizamos la operación del IMC y las posibles soluciones dependiendo del IMC del usuario.


def calculadora_imc():
    print('Esto es una calculadora de IMC, recuerda no usar números negativos o letras')

    peso = (input('Pon tu peso en Kilogramos (Kg): '))
    peso = peso.replace(',','.')
    peso = float(peso)
    print(f'Tu peso es de: {peso} Kg (ej: 82.1):')
    altura = (input('Pon tu altura en metros (m)(ej: 1.78): '))
    altura = altura.replace(',','.')
    altura = float(altura)
    print(f'Tu altura es de: {altura} m')
    
    IMC = peso/(altura**2)
    print(f'Tu IMC es: {IMC:.2f}')
    if IMC < 18.5:
        print('Estas bajo de peso')
    elif IMC< 25:
        print('Tu peso es normal')
    elif IMC< 30: 
        print('Tienes sobrepeso')
    else:
        print('Tienes obesidad')
        
calculadora_imc()
```

**EXERCICI 3**

Comptador de paraules d'un text.

Escriu una funció que donat un text, mostri les vegades que apareix cada paraula.

*Manera de hacer el ejercicio sin bibliotecas*


```python
# Se crea una función para borrar todos los signos de puntuación posibles y se reemplazan por un espacio en blanco, está función la llamaremos
# más adelante.

def remover_puntuacion(texto): 
    signos_puntuacion = [".", ",", "?", "¿", "!", "¡", "(", ")", ";", ":", "-", "'", '"']
    for signo in signos_puntuacion:
        texto = texto.replace(signo, ' ')
    return texto

# Esta función se crea para contar las palabras repetidas del texto introducido por el usuario.
# Esta función convierte el texto a una lista de cada palabra del texto por separado en minúsculas, para después añadir esas palabras a un diccionario
# y que nos cuente cuantas veces se repite esa palabra como se puede ver en el for.
def contador_veces(texto):
    texto = remover_puntuacion(texto) #aquí llamamos a la función anterior para introducir el texto sin ningún signo.
    lista_palabras = texto.lower().split()
    frecuencia = {}
    for palabra in lista_palabras:
        frecuencia[palabra] = lista_palabras.count(palabra)
    return frecuencia

texto = input('Escribe un texto: ')
resultado = contador_veces(texto)


# Aquí se muestra el resultado de una manera más clara, la palabra y el número de veces que aparece.
for palabra,numero in resultado.items():
    print(f'{palabra}:{numero}')
```

**EXERCICI 4**

Diccionari invers.

Resulta que el client té una enquesta molt antiga que s'emmagatzema en un diccionari i els resultats els necessita al revés, és a dir, intercanviats les claus i els valors. Els valors i claus en el diccionari original són únics; si aquest no és el cas, la funció hauria d'imprimir un missatge d'advertiment.

```python
diccionario = {1: 'a',
               2: 'b', 
               3: 'c', 
               4: 'b'}  

# Primero hacemos una lista con los valores del diccionario.
valores = list(diccionario.values())

# Luego hacemos un set de valores porque al hacer set, si hay valores duplicados automáticamente los borra.
valores_unicos = set(valores)

# Comparamos cuántos valores tiene las lista y el set, si son diferentes quiere decir que el set ha borrado valores duplicados.
if len(valores) != len(valores_unicos):
    print('¡Error! En el diccionario tienes valores duplicados')

# Crearemos un diccionario invertido en el que no se incluyan los valores duplicados.
# Si el valor no es duplicado, lo añadira al diccionario invertido y si es duplicado lo añadira el valor valores_vistos para poder sacar el error.
diccionario_invertido = {}
valores_vistos = set()

for clave, valor in diccionario.items():
    if valor in valores_vistos:
        print(f'El valor "{valor}" de la clave "{clave}" es duplicado, no se incluirá en el diccionario invertido.')
    else:
        diccionario_invertido[valor] = clave
        valores_vistos.add(valor)

print(diccionario_invertido)
```

```python
lista_cliente = [ '1.3', 'one' , '1e10' , 'seven', '3-1/2', ('2', 1, 1.4, 'not-a-number'), [1, 2, '3', '3.4']]

lista_float = []
lista_no_float = []

# Hacemos un for para los datos de la lista, para a continuación distinguir entre los que son tuplas o listas o los que no.
# Y ejecutamos si se puede convertir en float que lo añada a lista_float y si no se puede convertir que lo añada a la lista_no_float
for dato in lista_cliente:
    if isinstance(dato,(list, tuple)):
        for x in dato:
            try:
                float_value = float(x)
                lista_float.append(float_value)
            except ValueError:
                lista_no_float.append(x)
    else:
        try:
            float_value = float(dato)
            lista_float.append(float_value)
        except ValueError:
            lista_no_float.append(dato)

print(f'Esta es la lista con los datos que pueden convertir a float:\n {lista_float}')
print(f'Esta es la lista con los datos que no se pueden convertir a float:\n {lista_no_float}')



```
